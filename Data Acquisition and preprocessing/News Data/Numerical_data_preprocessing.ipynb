```python
def download_ticker(tick, start_date, end_date):
    
    filepath = './'+tick+'.csv'
  
    #Check if file exists
    
    check_file = os.path.exists(filepath)
    if check_file == True:
        print(check_file)
       
    else:
        
        tickA = yf.download(tick,start_date, end_date)    
        pd.DataFrame(tickA).to_csv(filepath)
        
l=['MSFT']   
```

```python
for item in l:
    download_ticker(item,"2011-01-01", "2020-04-30")
```

```python
price = []
for i in l:
    com1 = pd.read_csv(i + '.csv')
    sp1 = pd.DataFrame(com1)
    sp1['Date'] =  pd.to_datetime(sp1['Date'], format='%Y-%m-%d')
    sp1.set_index('Date', inplace=True)
    sp1 = sp1.resample('D').ffill().reset_index()
    sp1['day_of_week'] = sp1['Date'].dt.day_name()
    for index, row in sp1.iterrows():
        if row['day_of_week'] == 'Saturday':
            mon = sp1.iloc[index-1,2:7]
            fri = sp1.iloc[index+2,2:7]
            avg = (mon + fri) / 2
            sp1.iloc[index,2:7] = avg
            sp1.iloc[index+1,2:7] = avg
    price.append(sp1['Close'].to_numpy())
```
