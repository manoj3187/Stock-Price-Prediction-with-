```python
sp_500 = pd.read_csv('SP_500.csv')
```

```python
sp = pd.DataFrame(sp_500, columns = ["Ticker","Name"])
sp_dict = sp.to_dict()
sp.head()
```

```python
i = 0
l = []
while i < len(sp_dict['Ticker']):
    a = sp_dict['Ticker'][i]
    i += 1
    l.append(a)
l
```

```python
def download_ticker(tick, start_date, end_date):
    
    filepath = './'+tick+'.csv'
  
    #Check if file exists
    
    check_file = os.path.exists(filepath)
    if check_file == True:
        print(check_file)
       
    else:
        
        tickA = yf.download(tick,start_date, end_date)    
        pd.DataFrame(tickA).to_csv(filepath)
        
   
```

```python
for item in l:
    download_ticker(item,"2011-01-01", "2020-01-01")
```

```python
import pandas as pd
news= pd.read_csv('Microsoft.csv')
com1 = pd.read_csv('MSFT.csv')
sp1 = pd.DataFrame(com1)
sp2 = pd.DataFrame(news)
```

```python
len(sp1)
```

```python
len(sp2)
```

```python
sp1['Date'] =  pd.to_datetime(sp1['Date'], format='%Y-%m-%d')
sp1.set_index('Date', inplace=True)
sp1 = sp1.resample('D').ffill().reset_index()
sp1.head(30)
```

```python
sp1['day_of_week'] = sp1['Date'].dt.day_name()
sp1.head(30)
```

```python
sp1.iloc[1,2:7].shape
```

```python
for index, row in sp1.iterrows():
    if row['day_of_week'] == 'Saturday':
        mon = sp1.iloc[index-1,2:7]
        fri = sp1.iloc[index+2,2:7]
        avg = (mon + fri) / 2
        sp1.iloc[index,2:7] = avg
        sp1.iloc[index+1,2:7] = avg
sp1.head(30)
```

```python
sp2['date'] =  pd.to_datetime(sp2['date'], format='%Y-%m-%d')
```

```python
sp1 = sp1.rename(columns={'Date': 'date'})
```

```python
from functools import reduce
dfs = [sp1,sp2]
df_final = reduce(lambda left,right: pd.merge(left,right,on='date'), dfs)
```

```python
del df_final['Unnamed: 0']
```

```python
df_final.to_csv('sample.csv', header=True, encoding='utf-8')
```
